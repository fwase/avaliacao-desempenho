```python
import numpy as np
import random
import math
from queue import Queue
import matplotlib.pyplot as plt
```

```python
 # 1) Heurística MSER-5Y 

def mser(v,n=5):
  k = len(v)//n
  z = [sum(v[i:i+n])/n for i in range(0,len(v),n)]
  temp = []
  for i in range(k-1):
    temp.append(np.std(z[i::])/(k-i)**0.5) 
  print(temp) 
  d = temp.index(min(temp))
  if(d < k//2):
    print("Ponto de truncamento",d,"\nSobraram",len(v)-d*n,"elementos")
    return v[d*n::]
  else:
    print("Ponto de corte não encontrado", d)
    return v
  
```

```python
# 2) Executar o simulador da fila M/M/1. 

# Preliminares - variavel exponencial

def vaexp(v,b):
  return [-b*math.log(1-i) for i in v]

def va(b): 
  return vaexp([random.randint(1,2**32 - 1)/(2**32 - 1)],b)[0]

####### Simulação M/M/1 ############

def simula(n,f,mi):

  resultado = []
  fila = Queue()
  atual = 0

  for i in range(n):
    # tei - tempo entre intervalos.
    tei = va(1/f)
    # Se a fila estiver vazia.
    if fila.qsize() == 0:
      atual += tei
      fila.put(atual)
    else:
      #Se a fila estiver com itens.
      clk = atual
      while atual < clk + tei and fila.qsize()!=0:      
        resultado.append(atual-fila.get())
        atual += va(1/mi)
      if atual < clk + tei:
         atual = clk + tei
         fila.put(atual)
      else:        
        fila.put(clk + tei)
  while fila.qsize()!=0:    
    resultado.append(atual-fila.get())
    atual += va(1/mi)
  return resultado
```

3) Para cada um dos seguintes pares de valores de λ e μ, coletar N valores da métrica tempo médio de espera na fila (N=1.000, 2.000, 4.000, 8.000, 10.000) e determinar o fim do transiente utilizando o MSER-5Y 

a) λ = 8, μ = 10

```python
amostra = [1000, 2000,4000,8000,10000]
r = []
for i in amostra:
  r.append(simula(i,8,10))
print(r)
```

```python
r[0] = mser(r[0])
plt.title("n = 1000")
plt.plot(range(len(r[0])),r[0],"-k")
plt.show()
```
